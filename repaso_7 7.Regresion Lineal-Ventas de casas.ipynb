## <font color="#004D7F">2. Regresion Lineal</font>

### Concepto de Regresion Lineal

En estadística, la regresión lineal o ajuste lineal es un modelo matemático usado para aproximar la relación de dependencia entre una variable dependiente $y$, m variables independientes $X_i$ y un término aleatorio $\epsilon$.

La regresión lineal estima los coeficientes de la ecuación lineal, con una o más variables independientes, que mejor prediga el valor de la variable dependiente. Por ejemplo, puede intentar predecir el total de ventas anuales de un vendedor (la variable dependiente) a partir de variables independientes tales como la edad, la formación y los años de experiencia.

- Los modelos de regresión lineal no tienen en cuenta los espacios.
- Estadística Lineal: busca el modelamiento de una variable dependiente en función a una (Regresión Lineal Simple) o varias variables independientes (Regresión Lineal Múltiple)

Supuestos:
- Linealidad.
- Normalidad de los Datos.
- No Multicolinealidad (variables independientes relacionadas)
- Homocedasticidad

## <font color="#004D7F">2.1 Regresion Lineal Simple</font>

### Concepto : Regresión Lineal Simple

El objetivo de la regresión lineal simple (**univariada**) es modelar una relación entre una característica simple (**variable explicativa** $x$) y una **respuesta** de valor continuo (**variable TARGET** $y$). La ecuación se define:
$$y =  w_{0} + w_{1}*x$$
En este caso, el peso $w_{0}$ representa el eje $y$ interceptado y $w_{1}$ es el coeficiente ponderado de la variable explicativa. Nuestro objetivo es aprender los pesos de la ecuación lineal para describir la relación entre la variable explicativa y la variable destino, que después se pueden utilizar para predecir las respuestas de nuevas variables explicativas que no forman parte del conjunto de datos de entrenamiento.

Basándose en la ecuación lineal que hemos definido anteriormente, la regresión lineal se puede entender como la búsqueda de la recta de **mejor ajuste** a través de puntos de muestra, como se aprecia en la siguiente imagen:

## <font color="#004D7F">2.2 Regresion Lineal Multiple</font>

### Concepto : Regresión Lineal Multiple

El caso especial de la regresión lineal con una variable explicativa que hemos presentado en la subsección anterior también se denomina **regresión lineal simple**. Evidentemente, también podemos generalizar el modelo de regresión lineal a las variables explicativas múltiples; este proceso se denomina **regresión lineal múltiple**:
$$y =  w_{0} x_{0} + w_{1} x_{1} + ... +   w_{m} x_{m} = \sum_{i=0}^m w_{i} x_{i} = w^T x$$
En este caso, $w_{0}$ es el eje $y$ que intercepta con $x_{0}=1$.

## <font color="#004D7F">5.1.3.1 MSE  </font>

### MSE (Mean Squared Error)
**Sensible a valores atipicos**<br>
El MSE es una métrica muy utilizada y muy simple este indica que la diferencia cuadrática entre el valor real y el valor estimado.

¿Qué representa realmente el MSE? Representa la distancia al cuadrado entre los valores reales y los valores estimados, realizamos el cuadrado para evitar la cancelación de los términos negativos.

$$MSE= \frac{1}{n} \sum_{i=1}^n (y^{(i)} - \hat{y}^{(i)})^2$$

**Ventajas del MSE**

- La gráfica del MSE es diferenciable, por lo que se puede utilizar fácilmente como función de pérdida.

**Desventajas del MSE**

- El valor que se obtiene después de calcular el MSE es una unidad de salida al cuadrado. Por ejemplo, la variable de salida está en metros (m), entonces después de calcular el MSE la salida que obtenemos está en metros al cuadrado.
- Si hay valores atípicos en el conjunto de datos, se penalizan más los valores atípicos y el MSE calculado es mayor. Así que, en resumen, no es robusto a los valores atípicos.

## <font color="#004D7F">5.1.3.2 RMSE  </font>

### RMSE (Root Mean Squared Error)
**sensible a valores atipicos**<br>
La raiz del error cuadrático medio (RMSE) es la raíz cuadrada de MSE. Se usa más comúnmente que MSE porque, en primer lugar, a veces el valor de MSE puede ser demasiado grande para compararlo fácilmente.

$$MSE= \sqrt(\frac{1}{n} \sum_{i=1}^n (y^{(i)} - \hat{y}^{(i)})^2)$$

**Ventajas del RMSE**
- El valor de salida que se obtiene está en la misma unidad que la variable de salida requerida, lo que facilita la interpretación de la pérdida.

**Desventajas del RMSE**
- No es tan robusto a los valores atípicos.

## <font color="#004D7F">5.1.3.3 MAE  </font>

### MAE (Mean Absolute Error)
**Es mas robusto para valores atipicos en comparacion MSE Y RMSE**<BR>
El error absoluto medio (MAE) es similar al error cuadrático medio (MSE). Sin embargo, en lugar de la suma del cuadrado del error en MSE, MAE está tomando la suma del valor absoluto del valor real vs el valor estimado.

$$MAE= \frac{1}{n} \sum_{i=1}^n \vert{y^{(i)} - \hat{y}^{(i)}}\vert$$

**Ventajas del MAE**

- El MAE que se obtiene está en la misma unidad que la variable de salida.
- Es más robusto a los valores atípicos.

**Desventajas de MAE**

- La gráfica del MAE no es diferenciable por lo que tenemos que aplicar varios optimizadores como el Descenso Gradiente que sí puede ser diferenciable

## <font color="#004D7F">5.1.3.4 MAPE  </font>

### MAPE (Mean Absolute Percentage Error)

MAPE da el error en términos de porcentaje es decir que porcentaje del valor real esta sesgado el valor estimado.

Mientras que el MAE te da un valor que puedes discutir si es aceptable o no la métrica nunca te da una pista de cuánto "error" es aceptable.
Por ejemplo, si tiene un error absoluto medio de 10 dólares, ¿es demasiado? ¿O es aceptable? esto depende de la escala de su valor real. Por ejemplo si tenemos un valor predicho de 100 y el valor real es de 150, el error porcentual absoluto es (150 - 100) / 150 que da 33%.

$$MAPE= \frac{1}{n} \sum_{i=1}^n \frac{{y^{(i)} - \hat{y}^{(i)}}}{y^{(i)}}$$

**Algunas características**

- Puede tener algunos problemas para tratar los errores cuando el valor real es 0. Cuando hay algunos ejemplos como éste en su muestra, se recomiendan otras métricas.
- El MAPE tiene algunos problemas con los valores cercanos a cero. Si se esperan estos valores en su objetivo, debería elegir otra métrica.

<a id="section51341"></a>
## <font color="#004D7F">5.1.3.4.1 codigo  </font>

```python
### Modelo con 4 variables (LSTATA, INDUS, NOX, RM)
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

```

## <font color="#004D7F">5.1.3.5 R2 score  </font>

### R2 score

R2 score es una métrica que indica el rendimiento de su modelo, no el sesgo que tienen las predicciones. Así, con la ayuda de R2 score tenemos un modelo de referencia para comparar vs un modelo minimo cosa que ninguna de las otras métricas proporciona. Entonces, básicamente, R2 score calcula en qué medida la línea de regresión es mejor que una línea media.

R2 score también se conoce como Coeficiente de Determinación o a veces también se conoce como Bondad de ajuste.
$$R^2= 1 - \frac{SSE}{SST}$$
Donde: 

**SSE** error de la suma cuadrada de la línea de regresión.

**SST** error de la suma cuadrada de la línea media.

- VIF = 1 : Las variables no están correlacionadas.
- VIF < 5 : Las variables tienen una correlación moderada y se pueden quedar en el modelo.
- VIF >5 : Las variables están altamente correlacionadas y deben desaparecer del modelo.

```python
### Variable RM
lm_n = smf.ols(formula="RM~LSTAT+INDUS+NOX", data = df1).fit()
rsquared_n = lm_n.rsquared
VIF = 1/(1-rsquared_n)
VIF
```

```python
### Variable LSTAT
lm_n = smf.ols(formula="LSTAT~RM+INDUS+NOX", data = df1).fit()
rsquared_n = lm_n.rsquared
VIF = 1/(1-rsquared_n)
VIF
```

```python
### Variable INDUS
lm_n = smf.ols(formula="INDUS~RM+LSTAT+NOX", data = df1).fit()
rsquared_n = lm_n.rsquared
VIF = 1/(1-rsquared_n)
VIF
```

```python
### Variable NOX
lm_n = smf.ols(formula="NOX~RM+LSTAT+INDUS", data = df1).fit()
rsquared_n = lm_n.rsquared
VIF = 1/(1-rsquared_n)
VIF
```

## <font color="#004D7F">6.4 Homocedasticidad</font>

El supuesto de homogeneidad de varianzas, también conocido como supuesto de homocedasticidad, considera que la varianza es constante (no varía) en los diferentes niveles de un factor, es decir, entre diferentes grupos. En los modelos de regresión lineal, esta condición de homocedasticidad suele hacer referencia a los errores (residuos) del modelo, es decir, que la varianza de los errores es constante en todas las predicciones.
