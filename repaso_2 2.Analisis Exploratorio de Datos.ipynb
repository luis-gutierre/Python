## <font color="#004D7F">2.5 Mapa de correlacion de variables</font>

**Variable target : ttf**<br>
**variables predictoras que tienen alta correlacion entre ellas,ademas se aÃ±adio la variable target 'ttf' : variables_final**


```python
import seaborn as sns
cm=np.corrcoef(df[variables_final].values.T)
sns.set(font_scale=1.5)
fig=plt.figure(figsize=(15,12))
hm=sns.heatmap(cm,cbar=True,annot=True,square=True,fmt='.2f',annot_kws={'size':12},yticklabels=variables_final,xticklabels=variables_final)
plt.title('mapa de correlacion entre variables')
plt.show()
```

## <font color="#004D7F">2.6 Funcion Grafica</font>

### Crear una funcion que permita dar 4 tipos de graficos
#### 1. grafico de histogramas
#### 2.grafico de caja
#### 3.grafico de dispersion


##### def explore_col(s,e)

```python
#s :  el nombre dela variable a grafica
# e: numero de motores a grafica
```

```python
def explore_col(s,e):
    fig = plt.figure(figsize=(20,8))
   
    sub1 = fig.add_subplot(221)#221
    sub1.set_title(s + "histograma")
    sub1.hist(df[s])
        
    sub2 = fig.add_subplot(222)
    sub2.set_title(s+"diagrama de caja")
    sub2.boxplot(df[s])
    
    sub4 = fig.add_subplot(223)
    sub4.set:title("dispersion ")
    sub4.set_xlabel("ttf")
    sub4.scatter(df["ttf"],df[s])
    
    plt.show()
```

### La 'variable s12' con  la variable target  'ttf'

```python
explore_col("s12",15)  
```
